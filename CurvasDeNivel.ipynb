```python
## Problema 6
import matplotlib.pyplot as plt
from sympy import symbols, Eq, cos
import sympy as sp
import numpy as np
from sympy.plotting import plot3d, plot_implicit
from sympy import init_printing

init_printing(use_latex='mathjax')

# Definir variables
x, y, z = symbols('x y z')

# Rango de niveles para las curvas de nivel
niveles = np.arange(1, 5, 0.5, dtype=float)

# Definir la funcin
f1 = 1 + (x**2 + y**2)

# Graficar en 3D
plot3d(f1, (x, -3, 3), (y, -3, 3))

# Generar ecuaciones para las curvas de nivel
ecuaciones1 = [Eq(f1, nv) for nv in niveles]

# Graficar las curvas de nivel
cn1 = plot_implicit(ecuaciones1[0], (x, -4, 4), (y, -4, 4), line_color='red', show=False)
for i, eqq in enumerate(ecuaciones1):
    cn_i = plot_implicit(eqq, (x, -4, 4), (y, -4, 4), line_color=f'C{i}', show=False)
    cn1.append(cn_i[0])

# Mostrar curvas de nivel
cn1.show()
```

```python
## Problema 7
from sympy import symbols, Eq, cos, sqrt, acos
from sympy.plotting import plot3d, plot_implicit
import numpy as np


x, y, z = symbols('x y z')

# Rango de niveles
niveles = np.arange(-2, 3, 1)

# Definir la funcin
f8 = sqrt(x**2 + y**2)

# Graficar en 3D
plot3d(f8, (x, -3, 3), (y, -3, 3))

# Generar ecuaciones para las curvas de nivel
ecuaciones8 = [Eq(f8, nv) for nv in niveles]

# Graficar las curvas de nivel
cn8 = plot_implicit(ecuaciones8[0], (x, -4, 4), (y, -4, 4), line_color='red', show=False)
for i, eqq in enumerate(ecuaciones8):
    cn_i = plot_implicit(eqq, (x, -4, 4), (y, -4, 4), line_color=f'C{i}', show=False)
    cn8.append(cn_i[0])

# Mostrar curvas de nivel
cn8.show()

```

```python

f9 = cos((x**2 + y**2))

z_despejado = x**2 + y**2 - acos(f9)

# Graficar en 3D
plot3d(z_despejado, (x, -3, 3), (y, -3, 3))

# Generar ecuaciones para las curvas de nivel
ecuaciones9 = [Eq(z_despejado, nv) for nv in np.linspace(-1, 1, 5)]

# Graficar las curvas de nivel
cn9 = plot_implicit(ecuaciones9[0], (x, -4, 4), (y, -4, 4), line_color='red', show=False)
for i, eqq in enumerate(ecuaciones9):
    cn_i = plot_implicit(eqq, (x, -4, 4), (y, -4, 4), line_color=f'C{i}', show=False)
    cn9.append(cn_i[0])

# Mostrar curvas de nivel
cn9.show()
```
